## RQ2. How do the size and type of AI pull requests relate to review work and merge results, when we group pull requests into simple size and file-type buckets?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 1: Load core tables from AIDev

```python
pr = pd.read_parquet(
    "hf://datasets/hao-li/AIDev/all_pull_request.parquet",
    engine="fastparquet"
)

repo = pd.read_parquet(
    "hf://datasets/hao-li/AIDev/all_repository.parquet",
    engine="fastparquet"
)

# Commit-level file details for each PR
commit_details = pd.read_parquet(
    "hf://datasets/hao-li/AIDev/pr_commit_details.parquet",
    engine="fastparquet"
)

# Review comments per PR (code review)
pr_review_comments = pd.read_parquet(
    "hf://datasets/hao-li/AIDev/pr_review_comments.parquet",
    engine="fastparquet"
)
```

### Step 2: Filter to AI-generated PRs and compute outcomes

```python
# Start from the full PR table
pr_ai = pr.copy()

# Keep only PRs with a non-null AI agent label
pr_ai = pr_ai[pr_ai["agent"].notna()]

# Keep PRs that have a clear final state
pr_ai = pr_ai[pr_ai["state"].isin(["merged", "closed"])]

# Parse timestamps
pr_ai["created_at"] = pd.to_datetime(pr_ai["created_at"], errors="coerce")
pr_ai["closed_at"] = pd.to_datetime(pr_ai["closed_at"], errors="coerce")
pr_ai["merged_at"] = pd.to_datetime(pr_ai["merged_at"], errors="coerce")

# Binary merge outcome
pr_ai["is_merged"] = (pr_ai["state"] == "merged").astype(int)

# Time-to-merge or time-to-close (in days)
end_time = pr_ai["merged_at"].fillna(pr_ai["closed_at"])
pr_ai["time_to_merge_days"] = (
    end_time - pr_ai["created_at"]
).dt.total_seconds() / (3600 * 24)

# (Optional) attach repo stars for later analysis if needed
pr_ai = pr_ai.merge(
    repo[["id", "stars"]],
    left_on="repo_id",
    right_on="id",
    how="left",
    suffixes=("", "_repo")
)
pr_ai.rename(columns={"stars": "repo_stars"}, inplace=True)
```

### Step 3: Compute PR size and dominant file type using commit_details

```python
# Step 3.1: Aggregate commit-level stats per PR

# For each PR, sum additions, deletions and changes across all commits.
pr_size = (
    commit_details
    .groupby("pr_id")
    .agg(
        total_additions=("additions", "sum"),
        total_deletions=("deletions", "sum"),
        total_lines_changed=("changes", "sum")  # you can also use additions+deletions
    )
    .reset_index()
)

print("Rows in pr_size:", len(pr_size))
print(pr_size.head())
```

```python
# Step 3.2: Derive a coarse file_type from filename (since dataset does not include file_type explicitly)

if "file_type" not in commit_details.columns:
    # 3.2.1 Extract file extension from the filename (text after the last dot)
    ext = (
        commit_details["filename"]
        .str.extract(r"\.([^.]+)$")[0]   # get extension without the dot
        .str.lower()
    )

    # 3.2.2 Map file extensions into high-level file_type categories
    ext_to_type = {
        # Source code
        "py": "code", "java": "code", "js": "code", "ts": "code",
        "cpp": "code", "c": "code", "cs": "code", "go": "code",
        "rb": "code", "php": "code", "rs": "code", "kt": "code",

        # Configuration / infrastructure
        "yml": "config", "yaml": "config", "json": "config",
        "ini": "config", "toml": "config", "xml": "config",
        "cfg": "config",

        # Documentation
        "md": "docs", "rst": "docs", "txt": "docs", "adoc": "docs",

        # Tests (very rough heuristic – often test files contain these suffixes)
        "spec": "test", "test": "test",
    }

    commit_details["file_type"] = ext.map(ext_to_type).fillna("other")

# Quick sanity check
print(commit_details[["pr_id", "filename", "file_type"]].head())
```

```python
# Step 3.3: Determine dominant file type per PR

# 3.3.1 For each (PR, file_type), sum total lines changed
ft_agg = (
    commit_details
    .groupby(["pr_id", "file_type"])
    .agg(total_lines=("additions", "sum"))
    .reset_index()
)

# 3.3.2 For each PR, find the file_type with the largest total_lines
idx = ft_agg.groupby("pr_id")["total_lines"].idxmax()

main_ft = (
    ft_agg
    .loc[idx, ["pr_id", "file_type"]]
    .rename(columns={"file_type": "main_file_type"})
)

print("Rows in main_ft:", len(main_ft))
print(main_ft.head())

# 3.3.3 Merge dominant file type back into the PR-level size table
pr_size = pr_size.merge(main_ft, on="pr_id", how="left")

# Defensive coding:
# If the merge somehow did not create a 'main_file_type' column
# (e.g., main_ft is empty), create it and set everything to 'other'.
if "main_file_type" not in pr_size.columns:
    # No file-type information available → fallback category
    pr_size["main_file_type"] = "other"
else:
    # Fill missing values with 'other'
    pr_size["main_file_type"] = pr_size["main_file_type"].fillna("other")

print("Columns in pr_size:", pr_size.columns.tolist())
print(pr_size.head())
```

### Step 4: Compute review comment counts using pr_review_comments

```python
# 4.1 Extract PR number from pull_request_url, e.g. ".../pulls/123"
pr_review_comments["pr_number"] = (
    pr_review_comments["pull_request_url"]
    .str.extract(r"/pulls/(\d+)")
    .astype(float)
)
```

```python
# 4.2 Count how many review comments per PR number
review_counts = (
    pr_review_comments
    .groupby("pr_number")
    .agg(n_review_comments=("id", "nunique"))
    .reset_index()
)
```

### Step 5: Merge PR info with size, file type, and review metadata

```python
# 5.1 Merge PRs with commit_details-based size + main_file_type
#     Join key: pr.id <-> commit_details.pr_id  (confirmed earlier)
pr_with_size = pr.merge(
    pr_size,
    left_on="id",
    right_on="pr_id",
    how="inner"
)

print("PRs merged with size/file type info:", len(pr_with_size))
```

```python
# 5.2 Merge review comment counts
#     Join key: PR number <-> extracted pr_number from review URL
pr_with_size = pr_with_size.merge(
    review_counts,
    left_on="number",
    right_on="pr_number",
    how="left"
)

# Replace missing review counts with 0
pr_with_size["n_review_comments"] = pr_with_size["n_review_comments"].fillna(0)
```

```python
# 5.3 Filter AI-generated PRs AFTER joining all metadata
pr_ai = pr_with_size[pr_with_size["agent"].notna()].copy()

# Keep only PRs that have reached a final state.
# In GitHub API, both merged and rejected PRs have state == "closed".
pr_ai = pr_ai[pr_ai["state"] == "closed"].copy()

# Convert timestamps
pr_ai["created_at"] = pd.to_datetime(pr_ai["created_at"], errors="coerce")
pr_ai["closed_at"] = pd.to_datetime(pr_ai["closed_at"], errors="coerce")
pr_ai["merged_at"] = pd.to_datetime(pr_ai["merged_at"], errors="coerce")

# Merge outcome flag
# A PR is considered merged if merged_at is not null.
pr_ai["is_merged"] = pr_ai["merged_at"].notna().astype(int)

# Compute time to merge (fallback to closed_at when unmerged)
end_time = pr_ai["merged_at"].fillna(pr_ai["closed_at"])
pr_ai["time_to_merge_days"] = (
    end_time - pr_ai["created_at"]
).dt.total_seconds() / (3600 * 24)

# Handle missing values
pr_ai["main_file_type"] = pr_ai["main_file_type"].fillna("other")
pr_ai["total_lines_changed"] = pr_ai["total_lines_changed"].fillna(0)

print("Final number of AI PRs after merging:", len(pr_ai))
```

### Step 6: Create size buckets and summarize RQ2 metrics

```python
# 6.1 Compute rank-based tertile buckets for PR size
pr_ai["size_rank"] = pr_ai["total_lines_changed"].rank(method="first")

pr_ai["size_bucket"] = pd.qcut(
    pr_ai["size_rank"],
    q=3,
    labels=["Small", "Medium", "Large"]
)
```

```python
# 6.2 Aggregate merge outcomes, review activity, and PR size
group_cols = ["size_bucket", "main_file_type"]

rq2_summary = (
    pr_ai
    .groupby(group_cols, observed=False)
    .agg(
        n_prs=("id", "size"),
        merge_rate=("is_merged", "mean"),
        median_time_to_merge=("time_to_merge_days", "median"),
        mean_review_comments=("n_review_comments", "mean"),
        median_lines_changed=("total_lines_changed", "median")
    )
    .reset_index()
)
```

```python
# Convert merge rate to %
rq2_summary["merge_rate_pct"] = rq2_summary["merge_rate"] * 100

rq2_summary.head()
```

```python
print(pr_ai["is_merged"].value_counts(dropna=False))
print(rq2_summary["merge_rate"].describe())
```

### Step 7: Visualizations for RQ2

```python
# 7.1 Merge rate by size bucket and file type
plt.figure(figsize=(12, 6))
sns.barplot(
    data=rq2_summary,
    x="size_bucket",
    y="merge_rate_pct",
    hue="main_file_type"
)

plt.title("Merge Rate by PR Size Bucket and Main File Type")
plt.xlabel("PR Size Bucket")
plt.ylabel("Merge Rate (%)")
plt.legend(title="Main File Type")
plt.tight_layout()
plt.show()
```

```python
# 7.2 Median time-to-merge
plt.figure(figsize=(12, 6))
sns.barplot(
    data=rq2_summary,
    x="size_bucket",
    y="median_time_to_merge",
    hue="main_file_type"
)
plt.title("Median Time-to-Merge by PR Size Bucket and Dominant File Type")
plt.xlabel("PR Size Bucket")
plt.ylabel("Median Time-to-Merge (days)")
plt.legend(title="Main File Type")
plt.tight_layout()
plt.show()
```

```python
# 7.3 Average review comments (review effort)
plt.figure(figsize=(12, 6))
sns.barplot(
    data=rq2_summary,
    x="size_bucket",
    y="mean_review_comments",
    hue="main_file_type"
)
plt.title("Average Review Comments by PR Size Bucket and Dominant File Type")
plt.xlabel("PR Size Bucket")
plt.ylabel("Average # of Review Comments")
plt.legend(title="Main File Type")
plt.tight_layout()
plt.show()
```
